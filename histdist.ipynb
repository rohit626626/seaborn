```python
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['size'])
```

```python
sns.distplot(tips_df['tip'])
```

```python
sns.distplot(tips_df['total_bill'])
```

```python
sns.distplot(tips_df['total_bill'],bins=10)
```

```python
sns.distplot(tips_df['total_bill'], bins=100)
```

```python
sns.distplot(tips_df['total_bill'], hist=False)
```

```python
sns.distplot(tips_df['total_bill'], kde=False)
```

```python
sns.distplot(tips_df['total_bill'], rug=True)
```

```python
sns.distplot(tips_df['total_bill'], fit=norm, kde=False)
```

```python
sns.distplot(tips_df['total_bill'],color='r')
```

```python
sns.distplot(tips_df['total_bill'],vertical=True)
```

```python
sns.distplot(tips_df['total_bill'],axlabel='Total amount')
```

```python
sns.distplot(tips_df['total_bill'], label='Total bill')
plt.title('Histogram of Bill')
plt.legend()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df['total_bill'], bins=bins)
plt.xticks(bins)
plt.title('Histogram of Total Bill')
plt.show()
```

```python

plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'r','edgecolor':'g', 'linewidth':5, 'linestyle':'--', 'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'r','edgecolor':'g', 'linewidth':5, 'linestyle':'--', 'alpha':0.9},
kde_kws={'color':'y', 'linewidth':5, 'linestyle':'--', 'alpha':0.9},
rug=True, rug_kws={'color':'b', 'linewidth':2, 'linestyle':'--', 'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df['total_bill'], hist_kws={'color':'r','edgecolor':'g', 'linewidth':5, 'linestyle':'--', 'alpha':0.9},
kde=False, fit=norm,
rug=True, rug_kws={'color':'b', 'linewidth':12, 'linestyle':'--', 'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))
sns.distplot(tips_df['total_bill'], bins=9, label='total_bill')
sns.distplot(tips_df['tip'], bins=9, label='tip')
sns.distplot(tips_df['size'], bins=9, label='size')
plt.legend()
```
